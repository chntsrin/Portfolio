# **Introduction**

การวิเคราะห์บุคลิกภาพของลูกค้าเป็นการวิเคราะห์โดยละเอียดเกี่ยวกับลูกค้าในอุดมคติของบริษัท ช่วยให้บริษัทหรือธุรกิจเข้าใจลูกค้าได้ดีขึ้น และช่วยให้ปรับเปลี่ยนผลิตภัณฑ์ตามความต้องการเฉพาะ พฤติกรรม ของลูกค้าประเภทต่าง ๆ ได้ง่ายขึ้น 

และการวิเคราะห์บุคลิกภาพและพฤติกรรมการซื้อของลูกค้ายังช่วยให้ธุรกิจปรับเปลี่ยนผลิตภัณฑ์ตามลูกค้าเป้าหมายจากกลุ่มลูกค้าประเภทต่างๆ ตัวอย่างเช่น แทนที่จะใช้จ่ายเงินเพื่อทำการตลาดผลิตภัณฑ์ใหม่ให้กับลูกค้าทุกรายในฐานข้อมูลของบริษัท บริษัทสามารถวิเคราะห์ได้ว่าลูกค้ากลุ่มใดมีแนวโน้มที่จะซื้อผลิตภัณฑ์มากที่สุด และทำการตลาดผลิตภัณฑ์เฉพาะกลุ่มนั้นเท่านั้น

# **Data Description**

*   Kidhome :
จำนวนสมาชิกในครอบครัวที่เป็นเด็กของลูกค้า

*   Teenhome :
จำนวนสมาชิกในครอบครัวที่เป็นวัยรุ่นของลูกค้า

*   Dt_Customer :
วันที่ลูกค้าสมัครเป็นสมาชิก

จำนวนวันหลังจากการใช้จ่ายครั้งล่าสุด

จำนวนเงินที่ลูกค้าซื้อเนื้อในช่วง 2 ปีที่ผ่านมา

# **Data Preperation, Cleansing, Validity and Wrangling** 

ตรวจสอบ null ในแต่ละ column ของ Data set และทำการ drop ในส่วนของแถวที่มีข้อมูลที่เป็น null  

```python
missingCol = customer_data.isnull().sum()
missingCol = missingCol[missingCol>0]
print('Column:', missingCol.index[0], 'having', missingCol.values[0], 'missing values')
```

```python
#Data Cleansing
customer_data_cleaned = customer_data.dropna()

#เช็ค Data หลังจากที่ทำการ clean data
missingCol = customer_data_cleaned.isnull().sum()
missingCol = missingCol[missingCol>0]
print('Customer Data after cleaned have', missingCol.values.sum(), 'missing values')
```



> สร้าง column ใหม่ โดยให้ชื่อว่า 'Age' เก็บอายุของลูกค้าที่เป็นสมาชิก ซึ่งเป็นข้อมูลที่อ้างอิงมาจาก column 'Year_Birth'



> นอกจากนี้เรายังพบว่าใน 'Marital_Status' และ 'Education' มีคำหลายคำที่มีความหมายคล้ายคลึงกันในคำจำกัดความ







ดังนั้นเราจึงจัดกลุ่มเข้าด้วยกันเพื่อให้การวิเคราะห์ง่ายขึ้น โดยจะจัดกลุ่มดังนี้


**Marital_Status**
*   Single : Divorced, Widow, Alone, Absurd และ YOLO
*   Partner : Together และ Married

**Education**
*   Undergraduate : Basic และ 2n Cycle
*   Graduate : Graduation
*   Postgraduate : Master และ PhD


```python
print("Total categories in the feature Marital_Status:\n", customer_data_cleaned.Marital_Status.value_counts(), "\n")
print("Total categories in the feature Education:\n", customer_data_cleaned.Education.value_counts())
```

```python
customer_data_cleaned.Marital_Status = customer_data_cleaned.Marital_Status.replace({"Divorced": "Single", 
                                                                                     "Together": "Partner",
                                                                                     "Married": "Partner", 
                                                                                     "Widow": "Single", 
                                                                                     "Alone": "Single", 
                                                                                     "Absurd": "Single",
                                                                                     "YOLO": "Single"})
customer_data_cleaned.Education = customer_data_cleaned.Education.replace({"Basic":"Undergraduate",
                                                                           "2n Cycle":"Undergraduate", 
                                                                           "Graduation":"Graduate", 
                                                                           "Master":"Postgraduate", 
                                                                           "PhD":"Postgraduate"})
print(customer_data_cleaned.Marital_Status.unique())
print(customer_data_cleaned.Education.unique())
```

**ขั้นตอนต่อไปจะเป็นการเพิ่มฟีเจอร์ใหม่ที่ทำให้เราสามารถเรียกใช้งานข้อมูลได้ง่ายขึ้น ประกอบไปด้วย**


*   'Spending' คือ ยอดรวมของการใช้จ่ายทั้งหมดสำหรับลูกค้าแต่ละรายที่ชื่อ
*   'Time_With_Company' คือ จำนวนเดือนที่ลูกค้าเป็นสมาชิกกับบริษัท
*   'Children' คือ จำนวนลูกของลูกค้าที่เป็นสมาชิกที่อยุ่ในรายชื่อ โดยมีการรวมข้อมูลในส่วนของ 'Kidhome' และ 'Teenhome'
*   'Has_Child' คือ ข้อมูลที่แสดงว่าลูกค้าที่เป็นสมาชิกที่อยุ่ในรายชื่อมีลูกหรือไม่
*   'Family_size' คือ จำนวนของสมาชิกภายในครอบครัวของลูกค้า
*   'TotalAcceptedCmp' คือ จำของการเข้าร่วมแคมเปญทั้งหมดของลูกค้า












```python
customer_data_cleaned['Spending'] = customer_data_cleaned.MntWines.copy() + customer_data_cleaned.MntFruits.copy() + customer_data_cleaned.MntMeatProducts.copy() + customer_data_cleaned.MntFishProducts.copy() + customer_data_cleaned.MntSweetProducts.copy() + customer_data_cleaned.MntGoldProds.copy()
customer_data_cleaned['Time_With_Company'] = pd.to_datetime(customer_data_cleaned.Dt_Customer, dayfirst = True, format = '%d-%m-%Y')
customer_data_cleaned['Time_With_Company'] = pd.to_numeric(customer_data_cleaned.Time_With_Company.dt.date.apply(lambda z: (date.today() - z)).dt.days, downcast = 'integer') / 30
customer_data_cleaned['Children'] = customer_data_cleaned.Kidhome.copy() + customer_data_cleaned.Teenhome.copy()
customer_data_cleaned['Has_Child'] = np.where(customer_data_cleaned.Children > 0, "Has Child", "No Child")
customer_data_cleaned['TotalAcceptedCmp'] = customer_data_cleaned['AcceptedCmp1'].copy() + customer_data_cleaned['AcceptedCmp2'].copy() + customer_data_cleaned['AcceptedCmp3'].copy() + customer_data_cleaned['AcceptedCmp4'].copy() + customer_data_cleaned['AcceptedCmp5'].copy() + customer_data_cleaned['Response'].copy()

temp = []
for i in customer_data_cleaned.index: 
    if customer_data_cleaned['Marital_Status'][i] == 'Partner':
        temp.append(2 + customer_data_cleaned['Children'][i])
    else:
        temp.append(1 + customer_data_cleaned['Children'][i])
customer_data_cleaned['Family_size'] = temp
```
